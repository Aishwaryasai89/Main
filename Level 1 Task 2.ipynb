```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns",21)
```

```python
df = pd.read_csv('Dataset .csv')
df
```

```python
df["Restaurant Name"].value_counts()
```

```python
df.City.value_counts()
```

```python
city_restaurant_counts=df.groupby("City")["Restaurant Name"].count()
city_restaurant_counts
```

```python
city_with_highest_restaurant_counts=city_restaurant_counts.idxmax()
highest_restaurants_counts=city_restaurant_counts.max()
```

```python
print(f"The City with highest number of restaurant is'{city_with_highest_restaurant_counts}' with {highest_restaurants_counts} restaurants")
```

```python
average_ratings_by_city = df.groupby("City")["Aggregate rating"].mean()
```

```python
print("Average rating for restaurants in each city: \n",average_ratings_by_city)
```

```python
print(f"The City with the highest average rating is'{average_ratings_by_city.idxmax()}'with an average rating of '{average_ratings_by_city.max()}'")
```
